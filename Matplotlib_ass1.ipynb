What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

-->Represents individual data points and is useful for visualizing the distribution and relationship between two variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y, c='g',alpha=0.8)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('x and y plot')
```

Why is the subplot() function used? Draw four line plots using the subplot() function.


The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It is particularly useful when you want to display multiple plots in a single figure

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('x1&y1 plot')
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('x1&y1 plot')
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('x1&y1 plot')
plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('x1&y1 plot')
```

What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
-->Displays data using rectangular bars and is commonly used to compare values between different categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profits')
plt.title('profits')
```

```python
plt.barh(company,profit)
plt.xlabel('profits')
plt.ylabel('company')
plt.title('profits graph')
```

What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It displays key statistical measures such as the median, quartiles, and potential outliers. The box plot is useful for understanding the central tendency, spread, and skewness of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2],labels=['box1','box2'])

plt.title('Box Plot')
plt.xlabel('Datasets')
plt.ylabel('Values')

plt.show()
```
